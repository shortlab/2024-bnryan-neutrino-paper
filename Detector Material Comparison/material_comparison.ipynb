```python
import csv
import matplotlib.pyplot as plt
```

```python
# Import Data
aluminum_energies = []
aluminum_xs = []

with open('aluminum_xs.csv') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader :
        aluminum_energies.append(float(row[0])*10**(-3))
        aluminum_xs.append(float(row[1]))
```

```python
zinc_energies = []
zinc_xs = []

with open('zinc_xs.csv') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader :
        zinc_energies.append(float(row[0])*10**(-3))
        zinc_xs.append(float(row[1]))
```

```python
tin_energies = []
tin_xs = []

with open('tin_xs.csv') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader :
        tin_energies.append(float(row[0])*10**(-3))
        tin_xs.append(float(row[1]))
```

```python
# Total Comparison
plt.plot(zinc_energies, zinc_xs, label='Zinc')
plt.plot(aluminum_energies, aluminum_xs, label='Aluminum')
plt.plot(tin_energies, tin_xs, label='Tin')
plt.legend()
plt.xlabel('Neutrino Energy (keV)')
plt.ylabel('Cross Section (cm^2)')
plt.yscale('log')
plt.savefig('mat_comp_big.pdf')
```

```python
plt.plot(zinc_energies[0:100], zinc_xs[0:100], label='Zinc')
plt.plot(aluminum_energies[0:101], aluminum_xs[0:101], label='Aluminum')
plt.plot(tin_energies[0:95], tin_xs[0:95], label='Tin')

plt.xlabel('Neutrino Energy (keV)')
plt.ylabel('Cross Section (cm^2)')
plt.legend()
plt.yscale('log')
plt.savefig('mat_comp_small.pdf')
```
